```python
import uadapy.data as data
import uadapy.dr.uapca as uapca
import uadapy.plotting.plots_1d as plots_1d
```

# Loading data
We provide ready-to-use toy datasets for getting started with the analysis quickly. In this example, we fit normal distributions to the points of the iris-dataset.

```python
distribs_hi = data.load_iris_normal()
```

# Applying UAPCA
Uncertainty-aware principal component analysis can be applied directly to reduce the data to a user-defined target dimension, in this case 1D data.

```python
distribs_lo = uapca(distribs_hi, n_dims=2)
```

# Visualizing data
The data can be directly visualized, for example as boxplot.

```python
fig, axs = plots_1d.generate_multidim_boxplot(distribs_lo, 1000, dim_labels=["Dimension 1", "Dimension 2"])
```
