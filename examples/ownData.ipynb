```python
import numpy as np
import uadapy as ua
import uadapy.plotting.plots2D as plots2D
import uadapy.plotting.plotsND as plotsND
from sklearn import datasets
import uadapy.dr.uapca as uapca
```

# Working with own datasets
In this example, we want to show how it is possible to work on own datasets and create distributions based on them. Here, we use the wine dataset as an example. The data can be either modelled as normally distributed or using a KDE.

```python
wine = datasets.load_wine()
dist = []
for c in np.unique(wine.target):
    dist.append(ua.distribution(np.array(wine.data[wine.target == c]), "Normal"))
```

These distributions can be directly used in other functionality of the library.

```python
distribs_lo = uapca(dist, dims=2)
plots2D.plot_contour(distribs_lo)
```
