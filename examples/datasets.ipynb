```python
import uadapy
from uadapy import Distribution
import uadapy.data as data
import uadapy.dr.uapca as uapca
import uadapy.dr.wgmm_uapca as wgmm_uapca
from uadapy.plotting.plots_2d import plot_contour, plot_contour_bands, plot_samples
from uadapy.plotting.plots_nd import plot_contour as plot_contour_nd, plot_samples as plot_samples_nd
```

## Iris dataset as class-conditional distributions (KDE-based)

```python
distribs_hi = data.load_iris()
distribs_hi = [Distribution(uadapy.distributions.gmm_from_kde(d.kde)) for d in distribs_hi]
distribs_lo = wgmm_uapca(distribs_hi, n_dims=2)
fig, axs = plot_contour_bands(distribs_lo, show_plot=True)
```

## Iris dataset as class-conditional multivariate Normal distributions

```python
distribs_hi = data.load_iris_normal()
distribs_lo = uapca(distribs_hi, n_dims=2)
fig, axs = plot_contour_bands(distribs_lo, show_plot=True)
```
