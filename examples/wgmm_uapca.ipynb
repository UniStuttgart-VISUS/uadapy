```python
import numpy as np
import uadapy.data as data
import uadapy.dr.wgmm_uapca as wgmm_uapca
import uadapy.plotting.plots_2d as plots_2d
import uadapy.plotting.plots_1d as plots_1d
```

# Loading data
We provide a ready-to-use toy dataset for getting started with the analysis quickly. In this example, we use synthetic Gaussian Mixture Model distributions with multiple components per class.

```python
distribs_hi = data.generate_synthetic_gmm(n_classes=3, n_dims=4, random_state=18)
```

# Applying wGMM-UAPCA
Weighted Gaussian Mixture Model uncertainty-aware principal component analysis can be applied directly to reduce the GMM distributions to a user-defined target dimension. In this case, we reduce from 4D to 2D.

```python
weights = np.array([0.5, 0.3, 0.2])
distribs_lo = wgmm_uapca(distribs_hi, n_dims=2, weights=weights)
```

# Visualizing data
The data can be directly visualized, for example as a contour plot.

```python
plots_2d.plot_contour(distribs_lo, seed=0)
```
